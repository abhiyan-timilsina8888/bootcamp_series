```python
# copy and change the og array

import numpy as np
aee= np.array([1,2,3,4,5])
c=aee.copy()
aee[1]=0
print(aee)
print(c)
```

```python
#view and change the og array

import numpy as np
arr= np.array([1,2,3,4])
v= arr.view()
arr[0]= 80
print(arr)
print(v)
```

```python
# change the view and see the result 

import numpy as np
a= np.array([1,2,3])
v= a.view()
v[1]=100

print(a)
print(v)
```

```python
import numpy as np
arr= np.array([1,2,3,4])

x= arr.copy()
y= arr.view()

print(x.base) #-> returns none as it has it's own data
print(y.base) # -> returns the original array as it is a view of the original array

```
