```python
import numpy as np
arr= np.array([1,2,34], dtype= 'S')
print(arr)
print(arr.dtype)
```

```python
# forcing the data to change type

import numpy as np
arr= np.array([1.1,2.2,3.2])
n_arr=arr.astype('i') # int can also be used 64 bytes
print(n_arr)
print(n_arr.dtype)
```

```python
#integer to boolen

import numpy as np
arr = np.array([-1,1,2,3,0])
n_arr= arr.astype(bool)
print(n_arr)
print(n_arr.dtype)
```
