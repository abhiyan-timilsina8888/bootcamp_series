
# 1) uniform distributions
= discribes the probability where evry event has equal chance of occuring
# three parameters 
      low , high , size
      

```python
from numpy import random
x= random.uniform(low=0,high=1,size=(2,3))
print(x)
```

# visualization of uniform distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns


data = random.uniform(size=(2,3))

sns.displot(data)

plt.show()
```

# logistic distribution 
=.  describes the growth

# three paramters 
= loc, scale ,size -> loc=0 and scale =1 default 
 # note:
  =has same parameters as of normal distribution
  

```python
from numpy import random
x= random.logistic(loc=2,scale=3,size=(2,3))
print(x)
```

# visualization of logistic distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns


data = random.logistic(size=(1000))

sns.displot(data,kind='kde')

plt.show()
```

# difference between normal and logistic distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns


data={
    "normal":random.normal(loc=2,scale=4, size=1000),
    "logistic": random.logistic(loc=2,scale=3, size=1000)

}
sns.displot(data,kind='kde')

plt.show()
```

 # multinomial distribution
   = it is the genralization of binomial distrubution

# has three parameters
   n, pvals(list of probabilities of outcome) , size


```python
# sample for a dice roll
from numpy import random

x= random.multinomial(n=6 , pvals =[1/6,1/6,1/6,1/6,1/6,1/6])
print(x)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns


data = random.multinomial(n=6, pvals=[1/6,1/6,1/6,1/6,1/6,1/6], size=10)

sns.displot(data,kind='kde')

plt.show()
```

# exponential distribution
  = it describes the time until the next event.

# has two parameters :
   scale and size

```python
from numpy import random
x=random.exponential(scale=2 , size=(3,4))
print(x)
```

```python
# visualization of exponential distribution

from numpy import random
import matplotlib.pyplot as plt

import seaborn as sns

data= random.exponential(scale=2,  size=(3))
sns.displot(data,kind='kde')

plt.show()
```

# chi square distribution
== used as a basis to verify the hypothesis

# has 2 parameters
   df(degree of freesom) and size

```python
from numpy import random

x= random.chisquare(df =2 , size=(2,3))
print(x)
```

```python
#visualization of chi-square

from numpy import random

import matplotlib.pyplot as plt
import seaborn as sns

data= random.chisquare(df=2,size=100)

sns.displot(data, kind='kde')

plt.show()

```

# rayleigh distrubution
  == used in signal processing

# has two parameters

  scale ( standard deviation default 1.0) and size

  

```python
from numpy import random

x= random.rayleigh(scale=3, size=(2,3))
print(x)
```

```python
# visualization of raylieigh distribution

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data= random.rayleigh(scale=1, size=(100))

sns.displot(data,kind='kde')

plt.show()
```

# pareto distribution
 = based on pareto's law i.e 80-20 distribution (20% factor givee 80% outcome)

# has 2 parameters
  a=shape parameter and size 

```python
from numpy import random
x= random.pareto(a=2 , size=(1,9))
print(x)
```

```python
# visualization of pareto distribution

from numpy import random
import seaborn as sns
import matplotlib.pyplot as plt

data= random.pareto(a=2, size=100)

sns.displot(data , kind='kde')

plt.show()
```

# zipf distribution
= based on the zipf's law which states nth common term is the 1/n of the most common term.

# has two parameters 

  a= distribution parameter and size



```python
from numpy import random
import matplotlib.pyplot as plt

import seaborn as sns

data= random.zipf(a=2, size=(100))

sns.displot(data, kind='kde')

plt.show()
```
