```python
# shuffle() makes the changes in original array
from numpy import random
import numpy as np

arr= np.array([3,2,4,51])
random.shuffle(arr)
print(arr)
```

```python
#permutation() rearranges the original array and returns a new array while leaving the og arry unchanged

from numpy import random
import numpy as np

arr= np.array([3,4,7,1,9])
print(random.permutation(arr))
print(arr)
```
