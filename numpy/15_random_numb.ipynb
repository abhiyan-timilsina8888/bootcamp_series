```python
# generate a random number from 0 to 100
from numpy import random
x= random.randint(100)
print(x)
```

```python
#generate a float number between 0 to 1
from numpy import random
x= random.rand()
print(x)
```

```python
#generate a random array (1D)
from numpy import random
x= random.randint(100,size=(4))
print(x)
```

```python
#generate a random 2D array from 0 to 100 with 3 rows having 5 elemnts

from numpy import random
x= random.randint(100,size=(3,5))
print(x)
```

```python
#genration of random 2D float array
from numpy import random
x= random.rand(3,5)
print(x)
```

```python
#generate a random number from array

from numpy import random
x= random.choice([3,2,5,7])
print(x)
```

```python
from numpy import random
x= random.choice([3,5,9,7],size=(3,5))
print(x)
```
