```python
# concatenate()

import numpy as np
arr1= np.array([1,2,3])
arr2 = np.array([4,5,6])
arr= np.concatenate((arr1,arr2),axis=0)
print(arr)
```

```python
#join two 2d array

import numpy as np
arr1= np.array([[1,2,3],[2,3,4]])
arr2 = np.array([[3,2,3],[2,6,4]])
arr =np.concatenate((arr1,arr2),axis=1) 
print(arr)
```

note:
1) axis =0 denotes row-wise concatenation
2) axis =1 denotes column-wise concatenation

```python
#using stack for joining

import numpy as np
arr1= np.array([1,2,3])
arr2 = np.array([4,5,6])
arr = np.stack((arr1,arr2),axis=1)
print(arr)


```

```python
#use of hstack for horizontal stacking
import numpy as np
arr1= np.array([1,2,3,4])
arr2= np.array([9,8,7,6])
arr= np.hstack((arr1,arr2))
print(arr)
print(arr.ndim)
```

```python
#use of vstack for verticle stacking
import numpy as np
arr1= np.array([1,2,3,4])
arr2= np.array([9,8,7,6])
arr= np.vstack((arr1,arr2))
print(arr)
print(arr.ndim)
```

```python
#use of dstack for verticle stacking
import numpy as np
arr1= np.array([1,2,3,4])
arr2= np.array([9,8,7,6])
arr= np.dstack((arr1,arr2))
print(arr)
print(arr.ndim)
```
