binomial distribution is the discrete distribution.
three parameters n,p,size

```python
from numpy import random
x= random.binomial(n=10,p=0.5,size=10)
print(x)
```

```python
#visualization of binomial distribution
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random

sns.displot(random.binomial(n=10,p=0.5,size=100),kind='kde')
plt.show()
```

```python
#difference between normal and binomial distribution
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data={
    "normal": random.normal(loc=1,scale=2,size=1000),
   "binomial": random.binomial(n=10,p=0.5,size=1000)

}

sns.displot(data,kind='kde')
plt.show()

```
