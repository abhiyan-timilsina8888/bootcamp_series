# matplotlib(pyplot)- seaborn

#seaborn is a library that uses matplotlib underneath to plot graph i.e pyplot.

```python
#plotting a displot 

import matplotlib.pyplot as plt
import seaborn as sns

sns.displot([0,1,2,3,4,5],bins=4)
plt.show()
```

```python
# plotting a displot without histogram

import matplotlib.pyplot as plt
import seaborn as sns

sns.displot([0,1,2,3,4,5],kind='kde')
plt.show()
```

```python
#create a random normal distribution of size 2*3


from numpy import random

x= random.normal(size=(2,3))

print(x)
```

```python
# generate a random normal distribution with size,2*3 , mean=1, standard deviation=2

from numpy import random

x= random.normal(size=(2,3),loc=1,scale=2)

print(x)
```

```python
# Visualization of normal distribution

import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random

sns.displot(random.normal(size=1000),kind='kde')
plt.show()


```
